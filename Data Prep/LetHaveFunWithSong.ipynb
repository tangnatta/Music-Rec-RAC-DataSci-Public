```python
# Full

import json
import requests
from bs4 import BeautifulSoup
import time
import random

header = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

news = []

soup = BeautifulSoup(driver.page_source, 'html.parser')

for content in soup.find_all('div', {'class': 'css-7b225g e1e0ao7z2'}):
    for i in range(12):
        class_name = '__item_sectionnew_{} css-t8rc5b e11fmmx10'.format(i)
        new_section = content.find('div', {'class': class_name})
        att = new_section.find('a')
        title = att['title']
        link = att['href']

        release_time = new_section.find(
            'div', {'class': 'css-7wgfnr e11fmmx13'}).find('span').text
                
        cat = new_section.find(
            'div', {'class': 'css-7wgfnr e11fmmx13'}).find('a')
        category_title = cat['title']
        category_link = cat['href']

        try:
            time.sleep(random.random()*.05)
            err = False
            try:
                site = requests.get("https://www.thairath.co.th" + link, headers=header)
                
                # print("https://www.thairath.co.th" + link, ", status_code: " + str(site.status_code))
            except Exception as e:
                print("Requests Error: ",e)
                err = True
            
            if (site.status_code != 200):
                raise "Startus Code Error: " + str(site.status_code)
            
            site_soup = BeautifulSoup(site.content, 'html.parser')
            
            js_frame = site_soup.find_all(
                'script', {'type': 'application/ld+json', 'class': 'next-head'})[-1]

            js_frame = json.loads(js_frame.text)

            description = js_frame['description']

            article = js_frame["articleBody"]

            image = js_frame["image"]

            author = js_frame["author"]

            datePublished = js_frame["datePublished"]
            
            dateModified = js_frame["dateModified"]
        except Exception as e:
            print("Externded Part Error: ", e)
            err = True
        
        if (site.status_code != 200 or err == True):
            description = None

            article = None

            image = None

            author = None

            datePublished = None

            dateModified = None
        
        news.append({
            'title': title,
            'link': "https://www.thairath.co.th"+link,
            'description': description,
            'article': article,
            'image': image,
            'author': author,
            'category_title': category_title,
            'category_link': "https://www.thairath.co.th"+category_link,
            'datePublished': datePublished,
            'dateModified': dateModified,
            'time': release_time
        })
        # print(img)
        # print(link)

print(news)
print(len(news))
```
