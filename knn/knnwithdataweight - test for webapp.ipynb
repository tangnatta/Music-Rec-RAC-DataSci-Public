### User Data

```python
user = np.zeros((1, len(notes)))
user
```

```python
for i in range(len(notes)):
    user[0][i] = np.random.randint(0, 2)
user_df = pd.DataFrame(user, columns=notes)
user_df
```

```python
user_df.iloc[0, 3]
```

```python
freq[0]*(2**(-1/24))
```

```python
user_fq = []

for i in range(len(notes)):
    if user_df.iloc[0][i]!=0:
        min_fq = freq[i]*(2**(-1/24))
        max_fq = freq[i]*(2**(1/24))
        user_fq.append(np.random.uniform(min_fq, max_fq))
    else:
        user_fq.append(0.0)

user_fq
```

```python
user_df = pd.concat([user_df, pd.DataFrame(user_fq, index=notes).T], ignore_index=True)
user_df
```

```python
user_score = []

for i in range(len(notes)):
    if user_fq[i]!=0:
        if user_fq[i]>=freq[i]:
            n = 12*(math.log2(user_fq[i]/freq[i]))
        else:
            n = 12*(math.log2(freq[i]/user_fq[i]))

        user_score.append((1-n)*user_df.iloc[0][i])
    else:
        user_score.append(0.0)

user_score
```

```python
user_df = pd.concat([user_df, pd.DataFrame(user_score, index=notes).T], ignore_index=True)
user_df
```

```python
score_fq = []

for i in range(len(top5fq.index)):
    sumfq = 0
    for j in range(len(notes)):
        sumfq += user_df.iloc[2][j]*top5fq.iloc[i][j]

    score_fq.append(sumfq/numfreq[i]*100)


score_fq
```

```python
score_pc = []

for i in score_fq:
    score_pc.append(i/sum(score_fq)*100)

score_pc
```
